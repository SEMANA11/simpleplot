```python
russia_5940=russia.to_crs(5940)
# just Russia

cities_russia=cities[cities.COUNTRY=='Russia']
cities_5940=cities_russia.to_crs(russia_5940.crs)

rivers_5940=rivers.to_crs(russia_5940.crs)
rivers_5940=gpd.clip(gdf=rivers_5940,mask=russia_5940)
```

```python
base = russia_5940.plot(facecolor="khaki", edgecolor='black', linewidth=0.4,figsize=(5,5))
cities_5940.plot(marker='.', color='maroon', markersize=15,ax=base)
rivers_5940.plot(edgecolor='dodgerblue', linewidth=0.5,ax=base)
```

```python
# save the map as image
import matplotlib.pyplot as plt

base = russia_5940.plot(facecolor="khaki", edgecolor='black', linewidth=0.4,figsize=(5,5))
map2=cities_5940.plot(marker='.', color='maroon', markersize=15,ax=base)
mapEnd=rivers_5940.plot(edgecolor='dodgerblue', linewidth=0.5,ax=base)
plt.savefig(os.path.join("figures",'mapEnd.jpg'))
```

```python
worldMaps='https://github.com/SEMANA11/simpleplot/raw/main/maps/worldMaps.gpkg'
```

```python
countries=gpd.read_file(worldMaps,layer='countries')
cities=gpd.read_file(worldMaps,layer='cities')
rivers=gpd.read_file(worldMaps,layer='rivers')
```
